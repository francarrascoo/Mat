```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import fsolve

def v(t):
    agua_acumulada = (0.9*t*t) + 12.5*t + 36
    print (f"El agua acumulada en Valparaíso en {t} horas es de {agua_acumulada:.1f} milimetros")

def s(t):
    agua_acumulada = (0.95*t*t) + 16.8*t + 31
    print (f"El agua acumulada en Santiago en {t} horas es de {agua_acumulada:.1f} milimetros")

def c(t):
    agua_acumulada = (1.1*t*t) + 19.4*t + 26
    print (f"El agua acumulada en Concepcion en {t} horas es de {agua_acumulada:.1f} milimetros")

def rev_v(x):
    return 0.9*x*x + 12.5*x + 36 -100

def rev_s(x):
    return 0.95*x*x + 16.8*x + 31 -100

def rev_c(x):
    return 1.1*x*x + 19.4*x + 26 -100

while True:
    print ("1. Determinar el agua acumulada en Valparaíso respecto al tiempo transcurrido (en horas)")
    print ("2. Determinar el agua acumulada en Santiago respecto al tiempo transcurrido (en horas)")
    print ("3. Determinar el agua acumulada en Concepción respecto al tiempo transcurrido (en horas)")
    print ("4. Soluciones ejercicio 4")
    print ("5. Salir y mostrar gráfica")
    try:
        opcion = int(input("Ingrese opción: "))
        if opcion == 1:
            while True:
                try:
                    t = float(input("Ingrese horas trancurridas"))
                    v(t)
                    break
                except:
                    print("Ingrese número válido")
        elif opcion == 2:
            while True:
                try:
                    t = float(input("Ingrese horas trancurridas"))
                    s(t)
                    break
                except:
                    print("Ingrese número válido")
        elif opcion == 3:
            while True:
                try:
                    t = float(input("Ingrese horas trancurridas"))
                    c(t)
                    break
                except:
                    print("Ingrese número válido")
        elif opcion == 4:
            x = np.linspace(0,1000,1)
            solv = fsolve (rev_v, x)
            sols = fsolve (rev_s, x)
            solc = fsolve (rev_c, x)
            print (f"Valparaíso: {solv}")
            print (f"Santiago: {sols}")
            print (f"Concepción: {solc}")
            break
        elif opcion == 5:
            break
        else:
            print ("Ingrese opción válida")
    except:
        print ("Ingrese opción válida")

x_tiempo = np.arange(0,4,0.1)
y_valp = (0.9*x_tiempo*x_tiempo) + 12.5*x_tiempo + 36
y_san = (0.95*x_tiempo*x_tiempo) + 16.8*x_tiempo + 31
y_conce = (1.1*x_tiempo*x_tiempo) + 19.4*x_tiempo + 26
plt.xlabel("Tiempo transcurrido (en horas)")
plt.ylabel("Milimetros acumulados")
plt.title("Milimetros de agua acumulados respecto al tiempo (en horas) transcurrido")
plt.plot(x_tiempo, y_valp, color = 'm', ms =4, label = 'Valparaíso | 0.9*t*t + 12.5*t + 36 ')
plt.plot(x_tiempo, y_san, color = 'r', ms =4, label ='Santiago | 0.95*t*t + 16.8*t + 31' )
plt.plot(x_tiempo, y_conce, color = 'b', ms =4, label = 'Concepción | 1.1*t*t + 19.4*t + 26' )
plt.axis("tight")
plt.legend()
plt.grid(True)

#1. ¿Cuál sería la cantidad de agua acumulada en cada una de las ciudades al cabo de 40 minutos?
# - El agua acumulada en Valparapiso en 2/3 de hora (40 minutos) es de 44.7 milimetros
# - El agua acumulada en Santiago en 2/3 de hora (40 minutos) es de 42.7 milimetros
# - El agua acumulada en Concepción en 2/3 de hora (40 minutos) es de 39.4 milimetros

#2. Grafique las funciones respectivas.
# - Función 4 del menú

#3. Si la lluvia comenzó en Valparaíso a las 16:45 horas, en Santiago a las 17:45 horas y en Concepción a las 18:15 horas, 
# ¿cuál será el total de agua caída a las 20:00 horas en cada ciudad?
# - El agua acumulada en Valparaíso en 3.25 horas (3 horas y 15 minutos) es de 86.1 milimetros
# - El agua acumulada en Santiago en 2.25 horas (2 horas y 15 minutos) es de 73.6 milimetros
# - El agua acumulada en Concepcion en 1.75 horas (1 hora y 45 minutos) es de 63.3 milimetros

#4. ¿Después de cuánto tiempo cada ciudad requerirá por primera vez de la ayuda de los equipos de emergencia?

# - LLegarán cuando alcancen los 100mm por lo que:
# - En Valparaíso llegará luego de 3.98 horas
# - En Santiago llegará luego de 3.44 horas
# - En Concepción llegará luego de 3.225 horas

#5. Se espera que la lluvia dure 12 horas ininterrumpidas. Para gestionar posibles emergencias, se dispondrá de dos equipos de intervención:
#   - El primer equipo trabajará durante las primeras 6 horas de lluvia.
#   - El segundo equipo tomará el relevo durante las últimas 6 horas.
#   Basándote en el comportamiento de la acumulación de agua en las ciudades de Valparaíso, 
#   Santiago y Concepción. Analiza cómo cambia la acumulación de agua en cada ciudad a lo largo del día. 
#   ¿Cuál equipo tendrá que ir a revisar el alcantarillado con mayor frecuencia debido a la acumulación de agua? 
#   Justifica tu respuesta considerando la forma de las funciones que describen la acumulación de agua.
#   - Aunque teóricamente ninguna de las ciudades se inunde por el trabajo de los equipos, en Concepción 
#   se acumulará más agua en 6 horas, luego Santiago y luego Valparaíso de acuerdo a las funciones y lo que muestra el gráfico.
```
