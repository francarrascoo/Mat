## Graficar una función

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve   
import math

def f(x):
    return -1.6*x**2 + 1.3*x + 141.3

def y(x):
    return 1.4*x**2 - 5.6*x + 12.6

def inter(x):
    return f(x) - y(x)

sol = np.around(fsolve(inter, x),1);
sol_unica= np.unique(sol);
print (f"Las intersecciones respecto al eje x son: {sol_unica}")

def v(x):
    return  50*math.exp(0.3*x) - 551

xo= np.linspace(0, 10000, 1);
sol = fsolve(v, xo);
print (sol);

x = np.arange(-10, 15);
plt.plot(x, f(x), label = 'f(x)');
plt.plot(x, 1-2*x, label = 'y(x)');
plt.title('Nombre del gráfico');
plt.ylabel('Nombre eje y');
plt.xlabel('Nombre eje x');
plt.legend();
plt.grid();
plt.ylim(-15, 100);
plt.show();
```
