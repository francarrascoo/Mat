# Problema 1

Un equipo de desarrollo está evaluando la eficiencia de un nuevo algoritmo implementado en *Python*. Han medido el tiempo de ejecución del algoritmo en función de la cantidad de elementos de entrada, obteniendo los resultados que se muestran en la siguiente tabla:

<p><center>

| Cantidad de elementos (unidades) | Tiempo de ejecución (ms) |
|----------------------------------|--------------------------|
|$$ 100 $$                         | $$2 $$                   |
|$$ 200 $$                         | $$4 $$                   |
|$$ 500 $$                         | $$10 $$                  |
|$$ 1.000 $$                        | $$20 $$                  |
|$$ 2.000 $$                        | $$40 $$                  |

</center></p>

1. Defina variable dependiente e independiente, indicando unidad de medida.
2. Encuentra la función que relaciona el tamaño de la entrada con el tiempo de ejecución del algoritmo, para esto utiliza la función *polyfit* de *Numpy*.
3. Calcula el tiempo de ejecución para una entrada de $1.500$ elementos.
4. Utiliza la función *fsolve* de *Scipy* para determinar el tamaño de entrada que haría que el tiempo de ejecución sea de $50$ milisegundos.


```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import fsolve

cantidad_elementos = [100, 200, 500, 1000, 2000]
tiempo_ejecucion = [2, 4, 10, 20, 40]

m,n = np.polyfit(cantidad_elementos, tiempo_ejecucion, 1)
fun = f"{m:.2f}x + {n:.2f}"


def despeje_funcion(x,tiempo):
    return 0.02*x - tiempo

def calcular_elementos(tiempo):
    x = 0
    solucion = fsolve(despeje_funcion, x, args=(tiempo))
    return solucion[0]

while True:
    print ("1.- Encotrar función que relaciona las variables")
    print ("2.- Obtener el tiempo de ejecución en base a los elementos")
    print ("3.- Obtener los elementos en base al tiempo de ejecución")
    print ("4.- Salir y mostrar gráfica")
    try:
        opcion = int(input("Ingrese opción: "))
        if opcion == 1:
            print(f"La función es: f(x) = {fun}".replace("\n",""))
        elif opcion == 2:
            while True:
                try:
                    cantidad = int(input("Ingrese la cantidad de elementos: "))
                    funcion = 0.02*cantidad
                    print(f"El tiempo de ejecución para {cantidad} elementos es de {funcion} ms")
                    break
                except:
                    print("Cantidad inválida")
        elif opcion == 3:
            while True:
                try:
                    tiempo = float(input("Ingrese el tiempo de ejecución en milisegundos: "))
                    num_elementos = calcular_elementos(tiempo)
                    print(f"Al transcurrir {tiempo} ms, han ingresado {num_elementos} elemntos")
                    break
                except:
                    print("Tiempo inválido")
        elif opcion == 4:
            break
        else:
            print("Opción inválida")
    except:
        print("Opción inválida")

x = np.linspace(0, 2600)
y = 0.02*x

plt.xlabel("Cantidad de elementos")
plt.ylabel("Tiempo de ejecución (ms)")
plt.title("Tiempo de ejecución (en ms) en función de la cantidad de elementos")
plt.plot(x, y, color = 'm', ms =4)
plt.plot(cantidad_elementos, tiempo_ejecucion, 'o', color = 'c')
plt.grid(True)
plt.xlim([0, 2600]);
plt.ylim([0, 55]);

# 1.- Variable dependiente e independiente
#   - Variable dependiente: tiempo de ejecución en ms
#   - Variable independiente: cantidad de elementos en unidades
# 2.- Encontrar la función que relaciona ambas variables
#   - f(x) = 0.02 x
# 3.- Tiempo de ejecución para 1500 elementos
#   - f(1500) = 30 ms
# 4.- Elementos para un tiempo de ejecución de 50 ms
#   - 2500 elementos

```

# Problema 4

El uso de memoria $M(x)$ (en GB) de un servidor en función del número de usuarios activos $x$ está dado por la función lineal:

 $$M(x) = 0.5x + 2$$

1. Determina la variable dependiente e independiente, indicando unidad de medida.
2. Identifica e interpreta la pendiente de la función.
3. ¿Cuánta memoria del servidor está en uso cuando no hay usuarios activos?
4. ¿Cuánta memoria del servidor está en uso cuando hay $637$ usuarios activos?
5. Utiliza *Python* para determinar cuántos usuarios activos provocan un uso de memoria de $32$ GB.
6. ¿Es posible un uso de memoria de $59,8$ GB?

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import fsolve

def m(x):
    return 0.5*x + 2

def despeje_funcion(x,memoria):
    return 0.5*x + 2 - memoria

def calcular_usuarios(memoria):
    x = 0
    solucion = fsolve(despeje_funcion, x, args=(memoria))
    return solucion[0]

while True:
    print ("1.- Determinar la memoria utilizada en GB en función del número de usuarios activos")
    print ("2.- Determinar el número de usuarios activos en función de la memoria utilizada")
    print ("3.- Salir y mostrar gráfica")
    try:
        opcion = int(input("Ingrese opción: "))
        if opcion == 1:
            while True:
                try:
                    usuarios = int(input("Ingrese el número de usuarios activos: "))
                    print(f"La memoria utilizada es de {m(usuarios)} GB con {usuarios} usuarios activos")
                    break
                except:
                    print("Número inválido")
        elif opcion == 2:
            while True:
                try:
                    memoria = float(input("Ingrese la memoria utilizada en GB: "))
                    usuarios = calcular_usuarios(memoria)
                    print(f"Cuando la memoria utilizada es de {memoria} GB, el número de usuarios activos es de {usuarios}")
                    break
                except:
                    print("Memoria inválida")
        elif opcion == 3:
            break
        else:
            print("Opción inválida")
    except:
        print("Opción inválida")

x = np.linspace(0, 200)
y = m(x)

plt.xlabel("Usuarios activos")
plt.ylabel("Memoria utilizada (GB)")
plt.title("Memoria utilizada (en GB) en función del número de usuarios activos")
plt.plot(x, y, color = 'm', ms =4)
plt.grid(True)
plt.xlim([0, 120]);
plt.ylim([0, 65]);

#1.- Variable dependiente e independiente
#   -Variable dependiente: uso de memoria en GB
#   -Variable independiente: número de usuarios activos
#2.- La pendiente es 0.5 lo que indica que la función es creciente
#3.- La memoria utilizada cuando no hay usuarios activos es de 2 GB
#4.- La memoria utilizada cuando hay 637 usuarios activos es de 320.5 GB
#5.- Hay 60 usuarios activos cuando la memoria utilizada es de 32 GB
#6.- Como la cantidad de usuarios no puede ser decimal, no es posible un 
#    uso de memoria exacto de 59.8 GB. Cuando hay 115 usuarios activos el
#    consumo de memoria es de 59.5 GB y cuando hay 116 usuarios activos el
#    consumo de memoria es de 60 GB.
```

# Problema 5

El costo de operación mensual $C(x)$ (en miles de dólares) de un data center en función del número de servidores $x$ está dado por la función lineal:

 $$C(x) = 1,5x + 10$$

1. Determine la variable dependiente e independiente, indicando unidad de medida.
2. ¿Qué representa la pendiente en el contexto del problema?
3. ¿Qué representa el intercepto en el contexto del problema?
4. Calcula el costo de operación cuando hay $47$ servidores.
5. Determina cuántos servidores hay cuando el costo de operación es de $92.500$ dólares. Utiliza *Python* para resolver.


```python
import matplotlib.pyplot as plt
import numpy as np 
from scipy.optimize import fsolve

def c(x):
    return 1.5*x + 10

def despeje_funcion(x,costo):
    return 1.5*x + 10 - costo

def calcular_servidores(costo):
    x = 0
    solucion = fsolve(despeje_funcion, x, args=(costo))
    return solucion[0]

while True:
    print ("1.- Determinar el costo de operación mensual en función de los servidores activos")
    print ("2.- Determinar el número de servidores activos en función del costo de operación mensual")
    print ("3.- Salir y mostrar gráfica")
    try:
        opcion = int(input("Ingrese opción: "))
        if opcion == 1:
            while True:
                try:
                    servidores = int(input("Ingrese el número de servidores activos: "))
                    print(f"El costo de operación mensual es de {c(servidores)} mil dólares con {servidores} servidores activos")
                    break
                except:
                    print("Número inválido")
        elif opcion == 2:
            while True:
                try:
                    costo = float(input("Ingrese el costo de operación mensual en miles de dólares: "))
                    servidores = calcular_servidores(costo)
                    print(f"Cuando el costo de operación mensual es de {costo} mil dólares, el número de servidores activos es de {servidores}")
                    break
                except:
                    print("Costo inválido")
        elif opcion == 3:
            break
        else:
            print("Opción inválida")
    except:
        print("Opción inválida")

x = np.linspace(0, 200)
y = c(x)

plt.xlabel("Servidores activos")
plt.ylabel("Costo de operación mensual (miles de dólares)")
plt.title("Costo de operación mensual (en miles de dólares) en función del número de servidores activos")
plt.plot(x, y, color = 'm', ms =4)
plt.grid(True)
plt.xlim([0, 100]);
plt.ylim([0, 160]);

#1.- Variable dependiente e independiente
#   -Variable dependiente: costo de operación mensual en miles de dólares
#   -Variable independiente: número de servidores activos
#2.- La pendiente es 1.5 lo que indica que la función es creciente. Mientras más servidores, mayor costo.
#3.- El intercepto es 10, lo que indica que el costo fijo es de 10 mil dólares.
#4.- El costo de operación mensual es de 80.5 mil dólares cuando hay 47 servidores activos.
#5.- Hay 55 servidores activos cuando el costo de operación mensual es de 92.5 mil dólares.
```

# Problema 6

El número de visitas diarias $V(x)$ en un sitio web, en función del número de campañas publicitarias $x$, está modelado por la función:

$$V(x) = 300x + 500$$

Los administradores del sitio consideran contratar $20$ campañas publicitarias como máximo.

1. Defina variable dependiente e independiente del problema, indicando unidad de medida.
2. Determine el dominio contextualizado de la función.
3. Interprete la pendiente de la función en el contexto del problema.
4. Determine la cantidad de visitas diarias cuando se ejecutan $5$ campañas publicitarias.
5. Utilizando la función *fsolve* de *Scipy* responde: ¿cuántas campañas publicitarias se deben ejecutar para alcanzar $4.700$ visitas diarias?

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import fsolve

def v(x):
    return 300*x + 500

def despeje_funcion (x, cantidad_visitas):
    return 300*x + 500 - cantidad_visitas

def calcular_campañas(cantidad_visitas):
    x = 0
    solucion = fsolve(despeje_funcion, x, args=cantidad_visitas)
    return solucion[0]

while True:
    print ("1.- Determinar la cantidad de visitas diarias en función de las campañas publicitarias")
    print ("2.- Determinar la cantidad de campañas necesarias para alcanzar cierto número de visitas diarias")
    print ("3.- Salir y mostrar gráfica")
    try:
        opcion = int(input("Ingrese opción: "))
        if opcion == 1:
            while True:
                try:
                    cantidad_campañas = int(input("Ingrese cantidad de campaña(s) publicitarias: "))
                    if cantidad_campañas > 0 and cantidad_campañas <= 20:
                        print (f"La cantidad de visitas diarias con {cantidad_campañas} campaña(s) diarias es de {v(cantidad_campañas)}")
                        break
                    else:
                        print ("La cantidad de campañas debe ser mayor a 0 y menor o igual a 20")
                except:
                    print("Cantidad de campañas inválida")
        elif opcion == 2:
            while True:
                try:
                    cantidad_visitas = int(input("Ingrese cantidad de visitas: "))
                    if cantidad_visitas > 500 and cantidad_visitas <= 6500:
                        print (f"La cantidad de campañas publicitarias necesarias para alcanzar {cantidad_visitas} visitas diarias es de {calcular_campañas(cantidad_visitas)}")
                        break
                    else:
                        print ("La cantidad de visitas debe ser mayor a 500 y menor o igual a 6500")
                except:
                    print ("Cantidad inválida")    
        elif opcion == 3:
            break
        else:
            print ("Opción inválida")
    except:
        print ("Opción inválida")

x = np.linspace(0, 20)
y = v(x)

plt.xlabel("Cantidad de campañas publicitarias")
plt.ylabel("Cantidad de visitas diarias")
plt.title("Cantidad de visitas diarias en función de la cantidad de campañas publicitarias")
plt.plot(x, y, color = 'm', ms =4)
plt.grid(True)
plt.xlim([0, 20]);
plt.ylim([0, 6500]);

#1.- Variable dependiente e independiente
#   -Variable dependiente: Número de visitas diarias
#   -Variable independiente: Número de campañas publicitarias
#2.- Dom v(x) : [0, 20]
#4.- La cantidad de visitas diarias con 5 campañas diarias es de 2000
#5.- La cantidad de campañas publicitarias necesarias para alcanzar 4700 visitas diarias es de 14
```
