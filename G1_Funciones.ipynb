# Problema 1

Una red transmite datos a $100$ megabits por segundo. Crea una función utilizando *Python* que permita calcular la cantidad de datos transmitidos en una cierta cantidad de tiempo (en segundos). Luego utiliza la función y calcula cuántos datos se transmiten en:

* $45$ segundos
* $1,5$ minutos
* $1$ hora



```python
def f(x):
    y = 100*x
    print(f"Datos emitidos en {x} segundos: {y} megabits")

while True:
    print("1.- Calcular la cantidad de datos emitidos por un dispositivo en un tiempo determinado (en segundos)")
    print("2.- Salir")
    try:
        op = int(input("Seleccione una opción: "))
        if op == 1:
            while True:
                try:
                    x = int(input("Ingrese el tiempo en segundos: "))
                    f(x)
                    break
                except:
                    print("Por favor ingrese un número válido")
        elif op == 2:
            break
        else:
            print("Opción no válida")      
    except:
        print("Por favor ingrese un número válido")
```

# Problema 3

La **latencia** de una red corresponde al tiempo que tarda un paquete de datos en viajar desde el punto de origen al destino. En otras palabras, es el tiempo que transcurre desde que se envía una solicitud de un dispositivo hasta que se recibe una respuesta del servidor u otro dispositivo.

La latencia puede verse afectada por la velocidad de transmisión de los datos a través de los cables o conexiones inalámbricas, la distancia física entre los dispositivos, el tiempo que tarda un computador en procesar o reenviar los datos, etc.

En redes de comunicación, como las videoconferencias, juegos en línea y transmisión de datos en tiempo real, es deseable una baja latencia, de lo contrario pueden existir retrasos en la comunicación y afectar negativamente la experiencia del usuario.

Según el contexto mencionado, crea una función en *Python* que permita calcular la **latencia real** de una red dada una latencia estimada. Suponga que la latencia real es un $20\%$ mayor a la latencia estimada.

Calcule la latencia real para una latencia estimada de:

a)   $200$ milisegundos.

b)   $149$ milisegundos.

c)   $74$ milisegundos.



```python
def f(latencia_estimada):
    latencia_real = latencia_estimada*1.2
    print(f"Latencia real cuando la latencia estimada es {latencia_estimada}: {latencia_real} milisegundos")

while True:
    print("1.- Calcular la latencia real de un dispositivo dada la latencia estimada")
    print("2.- Salir")
    try:
        op = int(input("Seleccione una opción: "))
        if op == 1:
            while True:
                try:
                    latencia_estimada = float(input("Ingrese la latencia estimada: "))
                    f(latencia_estimada)
                    break
                except:
                    print("Por favor ingrese un número válido")   
        elif op == 2:
            break
        else:
            print("Opción no válida")
    except:
        print("Por favor ingrese un número")
        continue
```

# Problema 5

Un turista ha llegado a Santiago y desea conocer algunos lugares de la ciudad. Ha decidido visitar el Palacio de la Moneda y desde ahí trasladarse al centro comercial "Costanera Center", utilizando algún medio de transporte que ofrece la ciudad.

Si se traslada en metro deberá abordar en estación La Moneda y bajar en la estación Tobalaba ($9$ estaciones). La función $f(t)$ permite calcular la distancia recorrida utilizando el metro (en kilómetros) transcurridos $t$ minutos.

$$f(t) = 0,4t$$

Si se traslada en bus, el turista podrá observar la ciudad y otros atractivos en su viaje. La función $g(t)$ permite calcular la distancia recorrida en bus (en kilómetros) transcurridos $t$ minutos.

$$g(t) = 0,3t$$

Según la información anterior:

1. Grafique ambas funciones, indicando el nombre de cada eje junto con su unidad de medida. Para realizar el gráfico utilice la biblioteca *Matplotlib*.
2.  Si se sabe que el metro se demora $1,2$ minutos en llegar desde una estación a otra y espera $30$ segundos en cada estación, indique el dominio contextualizado para $f(t)$.
3. Mediante análisis gráfico, indique cuál medio de transporte es más conveniente en términos de tiempo, para el turista. Justifique.
4.  Si se sabe que desde estación La Moneda hasta Tobalaba son aproximadamente $6$ kilómetros ¿cuántos tiempo tardará el turista en llegar a su destino con cada una de las opciones?

```python
import matplotlib.pyplot as plt
import numpy as np

def metro(tiempo):
    distancia = 0.4*tiempo
    print (f"La distancia recorrida en metro en {tiempo} minutos es de {distancia:.2f} km")

def inversa_metro(distancia):
    tiempo = distancia/0.4
    print(f"Se necesitan {tiempo:.2f} minutos para recorrer {distancia} km en metro")

def bus(tiempo):
    distancia = 0.3*tiempo
    print (f"La distancia recorrida en bus en {tiempo} minutos es de {distancia:.2f} km")

def inversa_bus(distancia):
    tiempo = distancia/0.3
    print(f"Se necesitan {tiempo:.2f} minutos para recorrer {distancia} km en bus")

while True:
    print("1.- Calcular la distancia recorrida (en kilometros) en metro y bus en un tiempo determinado (en minutos)")
    print("2.- Calcular el tiempo en recorrer cierta distancia (en kilometros) en metro y bus")
    print("3.- Salir y mostrar gráfica")
    try:
        opcion = int(input("Seleccione una opción: "))
        if opcion == 1:
            while True:
                try:
                    tiempo = float(input("Ingrese el tiempo en minutos: "))
                    metro(tiempo)
                    bus(tiempo)
                    break
                except:
                    print("Por favor ingrese un número válido")
        elif opcion == 2:
            while True:
                try:
                    distancia = float(input("Ingrese la distancia en kilómetros: "))
                    inversa_metro(distancia)
                    inversa_bus(distancia)
                    break
                except:
                    print("Por favor ingrese un número válido")
        elif opcion == 3:
            break
    except:
        print("Por favor ingrese un número")
        continue

x_tiempo = np.linspace(0, 100)
y_distancia_metro = 0.4*x_tiempo
y_distancia_bus = 0.3*x_tiempo
plt.xlabel("Tiempo (minutos)")
plt.ylabel("Distancia (km)")
plt.title("Distancia recorrida en metro y bus en función del tiempo")
plt.plot(x_tiempo, y_distancia_metro, color = 'r', label = 'Metro | f(t) = 0,4t', ms =4)
plt.plot(x_tiempo, y_distancia_bus, color = 'm', label = 'Bus | g(t) = 0,3t', ms = 4)
plt.legend()
plt.grid(True)
plt.xlim([0, 25]);
plt.ylim([0, 10]);

# 2.- Dominio para f(t)
#     t∈[0,13.6]
# 3.- El metro es más conveniente para el turista ya que recorre una mayor distancia en menos tiempo
# 4.- Para reccorrer 6 km, en metro se tarda 15 minutos y en bus 20 minutos
```

$$T(t) = -0,5 t^{2} + 3 t + 20$$

```python
import matplotlib.pyplot as plt
import numpy as np

def temperatura(tiempo):
    temperatura = -0.5*tiempo*tiempo + 3*tiempo + 20
    print (f"La temperatura a las {tiempo} horas es de {temperatura:.2f}°C")

while True:
    print("1.- Calcular la temperatura de un servidor en función del tiempo (en horas)")
    print("2.- Salir y mostrar gráfica")
    try:
        opcion = int(input("Seleccione una opción: "))
        if opcion == 1:
            while True:
                try:
                    tiempo = float(input("Ingrese el tiempo en horas: "))
                    if tiempo < 0 or tiempo > 9:
                        print("El  tiempo no puede ser menor que cero ni mayor que 9 horas en este caso") 
                    else:
                        temperatura(tiempo)
                        break
                except:
                    print("Por favor ingrese un número válido")
        elif opcion == 2:
            break
    except:
        print("Por favor ingrese un número válido")
        continue

plt.xlabel("Tiempo (horas)")
plt.ylabel("Temperatura (°C)")
plt.title("Temperatura de un servidor en función del tiempo")
x_tiempo = np.linspace(0, 9)
y_temperatura = -0.5*x_tiempo*x_tiempo + 3*x_tiempo + 20 
plt.plot(x_tiempo, y_temperatura, color = 'm', ms =4)
plt.grid(True)
plt.xlim([0, 9]);
plt.ylim([0, 25]);

# 1.- Variables dependiente e independiente
#     Variable independiente: t (tiempo en horas)
#     Variable dependiente: T(t) (temperatura en grados Celsius)
# 2.- Dominio
#     t∈[0,9]
# 4.- Máxima temperatura
#     A las 11:00 horas (3h) alcanza su temperatura máxima de 24.5°C
# 5.- Temperatura a las 13:00 (5h) y 17:00 horas (9h)
#     A las 13:00 horas la temperatura es de 22.5°C y a las 17:00 horas es de 6.5°C
```

# Problema 7

El número de usuarios de una red social se puede modelar mediante la función:

$$U(t) = \frac{1.000}{1+9e^{-0.5t}}$$

donde $t$ corresponde al tiempo transcurrido en meses.

1.  Defina variable dependiente e independiente, indicando unidad de medida.
2.  Determine la cantidad de usuarios transcurridos $12$ meses.
3.  Grafique la función, utilizando la librería *Matplotlib*, para los primeros dos años de funcionamiento.
4.  ¿Cuánto tiempo debe pasar para que la red social llegue a $800$ usuarios?

```python
import matplotlib.pyplot as plt
import numpy as np

def usuarios(tiempo):
    usuarios = 1000/(1+9*np.exp(-0.5*tiempo))
    print (f"El número de usuarios en {tiempo} meses es de {usuarios:.2f}")

def inversa_usuarios(usuarios):
    tiempo = -2 * np.log((1000 / (9 * usuarios)) - (1 / 9))
    print (f"El tiempo en que se alcanzarán {usuarios} usuarios es de {tiempo:.2f} meses")

while True:
    print ("1.- Calcular el número de usuarios en un tiempo determinado (en meses)")
    print ("2.- Calcular el tiempo en que se alcanzarán cierto número de usuarios")
    print ("3.- Salir y mostrar gráfica")
    try:
        opcion = int(input("Seleccione una opción: "))
        if opcion == 1:
            while True:
                try:
                    tiempo = float(input("Ingrese el tiempo en meses: "))
                    if tiempo < 0 or tiempo > 24:
                        print("El tiempo no puede ser menor que cero ni mayor que 24 meses en este caso")
                    else:
                        usuarios(tiempo)
                        break
                except:
                    print("Por favor ingrese un número válido")
        elif opcion == 2:
            while True:
                try:
                    usuarios = float(input("Ingrese el número de usuarios: "))
                    if usuarios < 100 or usuarios >= 1000:
                        print("El número de usuarios debe ser mayor a 100 y menor a 1000")
                    else:
                        inversa_usuarios(usuarios)
                        break
                except:
                    print("Por favor ingrese un número válido")
        elif opcion == 3:
            break
    except: 
        print("Por favor ingrese un número")
        continue


x_tiempo = np.linspace(0, 24)
y_usuarios = 1000/(1+9*np.exp(-0.5*x_tiempo))
plt.xlabel("Tiempo (meses)")
plt.ylabel("Número de usuarios")
plt.title("Número de usuarios en función del tiempo")
plt.plot(x_tiempo, y_usuarios, color = 'm', ms =4)
plt.grid(True)
plt.xlim([0, 24]);
plt.ylim([0, 1000]);

# 1.- Variables dependiente e independiente
#     Variable independiente: t (tiempo en meses)
#     Variable dependiente: U(t) (número de usuarios)
# 2.- Cantidad de usuarios en 12 meses
#     Aproximadamente 978 usuarios
# 4.- Tiempo en que se alcanzarán 800 usuarios
#     Aporximadamente en 7.17 meses

```

# Problema 9

El consumo de energía (en $kWh$) de un *Data Center* puede ser modelado por la función:

 $$E(t) = 50\cdot \log(t + 1) + 200$$

donde $t$ corresponde al tiempo (en horas) desde el inicio del monitoreo.

1. Defina variable dependiente e independiente, indicando unidad de medida.
2. Determine el consumo de energía del *Data Center* después de $5$ horas.
3. ¿Luego de cuántas horas el consumo será de $350$ kWh?
4. Grafique, utilizando la biblioteca *Matplotlib*, la función $E(t)$.

```python
import matplotlib.pyplot as plt
import numpy as np
import time
from os import system

def limpiar_pantalla():
    system("cls")

def tiempo(cant_elementos):
    tiempo = 0.01*cant_elementos*cant_elementos + 0.5*cant_elementos + 2
    print (f"El tiempo de ejecución para {cant_elementos} elementos es de {tiempo:.2f} segundos")

def inversa_tiempo(tiempo):
    n1 = (-0.5 + np.sqrt(0.25 - 0.04*(2 - tiempo)))/0.02
    n2 = (-0.5 - np.sqrt(0.25 - 0.04*(2 - tiempo)))/0.02
    cant_elementos = max(n1, n2)
    horas = time.strftime('%H:%M:%S', time.gmtime(tiempo))
    print (f"La cantidad de elementos que se pueden ordenar en {tiempo} segundos ({horas} horas) es de {cant_elementos:.2f} elementos")

def conversion(horas):
    segundos = horas*3600
    print(f"{horas} horas equivalen a {segundos} segundos")

def conversion_segundos(segundos):
    conversion = time.strftime('%H:%M:%S', time.gmtime(segundos))
    print(f"{segundos} segundos equivalen a {conversion} horas")

while True:
    limpiar_pantalla()
    print("1.- Calcular el tiempo de ejecución para una cantidad de elementos")
    print("2.- Calcular la cantidad de elementos que se pueden ordenar en cierto tiempo")
    print("3.- Convertir horas a segundos")
    print("4.- Convertir segundos a horas")
    print("5.- Salir y mostrar gráfica")

    try:
        opcion = int(input("Seleccione una opción: "))
        if opcion == 1:
            limpiar_pantalla()
            while True:
                try:
                    cant_elementos = float(input("Ingrese la cantidad de elementos: "))
                    tiempo(cant_elementos)
                    break
                except:
                    print("Por favor ingrese un número válido")
        elif opcion == 2:
            limpiar_pantalla()
            while True:
                try:
                    tiempo = float(input("Ingrese el tiempo en segundos: "))
                    inversa_tiempo(tiempo)
                    break
                except:
                    print("Por favor ingrese un número válido")
        elif opcion == 3:
            limpiar_pantalla()
            while True:
                try:
                    horas = float(input("Ingrese las horas a convertir: "))
                    conversion(horas)
                    break
                except:
                    print("Por favor ingrese un número válido")         
        elif opcion == 4:
            limpiar_pantalla()
            while True:
                try:
                    segundos = float(input("Ingrese los segundos a convertir: "))
                    conversion_segundos(segundos) 
                    break
                except:
                    print("Por favor ingrese un número válido")
        elif opcion == 5:
            break
        else:
            print("Opción no válida")
        
        input("Presione una tecla para continuar...")
    except:
        print("Por favor ingrese un número")
        continue


x_cant_elementos = np.linspace(0, 1562)
y_tiempo = 0.01*x_cant_elementos*x_cant_elementos + 0.5*x_cant_elementos + 2
plt.xlabel("Cantidad de elementos")
plt.ylabel("Tiempo de ejecución (segundos)")
plt.title("Tiempo de ejecución en función de la cantidad de elementos")
plt.plot(x_cant_elementos, y_tiempo, color = 'm', ms =4)
plt.grid(True)
plt.xlim([0, 1562]);
plt.ylim([0, 26000]);

# 1.- Variables dependiente e independiente
#     Variable independiente: n (cantidad de elementos a ordenar)
#     Variable dependiente: A(n) (tiempo de ejecución en segundos)
# 2.- Dominio
#     n∈[0, 1562]
# 3.- Tiempo de ejecución para 1200 elementos
#     15002 segundos (4 horas, 10 minutos y 2 segundos)
# 5.- Elementos ordenados luego de 6 horas (21600 segundos)
#     Aproximadamente 1444 elementos
```

# Problema 10

En gestión de proyectos, es crucial entender cómo se distribuye la carga de trabajo a lo largo del tiempo para planificar recursos, tiempo y esfuerzos de manera eficiente. Un fenómeno común es la disminución exponencial de la carga de trabajo, donde el esfuerzo requerido es mayor al inicio del proyecto y disminuye gradualmente a medida que se completan las tareas principales.

Esta disminución exponencial puede modelarse matemáticamente para predecir y gestionar el trabajo de manera efectiva.

<br>

Un estudio sobre gestión de proyectos sostiene que la carga de trabajo en un proyecto (en porcentaje) puede ser modelada por la función:

$$W(t) = 100e^{-0.1t}$$

donde $t$ corresponde al tiempo transcurrido (en semanas) desde el inicio del proyecto.

1. Defina variable dependiente e independiente, indicando unidad de medida.
2. Determine la carga de trabajo al inicio del proyecto.
3. ¿Cuál será la carga de trabajo luego de transcurridas $4$ semanas desde el inicio del proyecto?
4. Si el proyecto duró $12$ semanas, ¿es correcto afirmar que la carga de trabajo llegó al $20\%$?
5. ¿Cuántas semanas han pasado desde el inicio del proyecto para que la carga de trabajo sea de $55\%$?
6. Utilizando *Matplotlib*, grafique la función considerando un tiempo máximo de $12$ semanas.

```python
import matplotlib.pyplot as plt
import numpy as np
from os import system

def limpiar_pantalla():
    system("cls")

def carga_trabajo(tiempo):
    carga=100*np.exp(-0.1*tiempo)
    print (f"La carga de trabajo a las {tiempo} semanas es del {carga:.2f}%")

def inverse_carga(carga):
    tiempo = -10*np.log(carga/100)
    print (f"Se necesitan {tiempo:.2f} semanas para alcanzar el {carga}% de carga de trabajo")

while True:
    limpiar_pantalla()
    print("1.- Calcular la carga de trabajo (en porcentaje) en un tiempo determinado (en semanas)")
    print("2.- Calcular el tiempo en alcanzar cierta carga de trabajo (en semanas)")
    print("3.- Salir y mostrar gráfica")
    try:
        opcion = int(input("Seleccione una opción: "))
        if opcion == 1:
            limpiar_pantalla()
            while True:
                try:
                    tiempo = float(input("Ingrese el tiempo en semanas: "))
                    carga_trabajo(tiempo)
                    break
                except:
                    print("Por favor ingrese un número válido")
        elif opcion == 2:
            limpiar_pantalla()
            while True:
                try:
                    carga = float(input("Ingrese la carga de trabajo en porcentaje: "))
                    inverse_carga(carga)
                    break
                except:
                    print("Por favor ingrese un número válido")
        elif opcion == 3:
            break
        else:
            print("Opción no válida")
        
        input("Presione una tecla para continuar...")
        
    except:
        print("Por favor ingrese un número")
        continue

x_tiempo = np.linspace(0, 12)
y_carga = 100*np.exp(-0.1*x_tiempo)
plt.xlabel("Tiempo (semanas)")
plt.ylabel("Carga de trabajo (%)")
plt.title("Carga de trabajo en función del tiempo")
plt.plot(x_tiempo, y_carga, color = 'm', ms =4)
plt.grid(True)
plt.xlim([0, 12]);
plt.ylim([0, 100]);

# 1.- Variables dependiente e independiente
#     Variable dependiente: w(t) carga de trabajo en porcentaje
#     Variable independiente: (t) tiempo en semanas
# 2.- Carga al inicio del proyecto
# 3.- Carga a las 4 semanas
#     67.03%
# 4.- Carga del proyecto si dura 12 semanas
#     30.12%
# 5.- Semanas transcurridas para alcanzar el 55% de carga
#     5.98 semanas
```
